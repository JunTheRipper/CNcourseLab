```python
import numpy as np
import pandas as pd
import os

def get_filename(number: int) -> str:
    try:
        for i in os.listdir("data"):
            keyword = str(number)+"_"
            if i[-3:] == 'csv' and i.startswith(keyword):
                filename = i # Only One!
        return filename
    except Exception:
        return 'NAN'

def data_processing(number: int) -> pd.DataFrame:
    '''
    para:
        number of subject
    '''
    k = get_filename(number)
    if get_filename(number) == 'NAN':
        return 'NAN'    
    else:
        filename = "data/"+ k 
    return filename

# raw_data = data_processing(432132)[1:]
# raw_data
def raw_data_exange(number: int) -> pd.DataFrame:
    flname = data_processing(number)
    x = ''
    with open(flname,'r+', encoding='utf-8') as file:
        x = file.read()
        x = x.replace('\n\n','').replace('，\n','，').replace('？\n','')
    file.close()

    with open(flname,'w', encoding='utf-8') as file2:
        file2.write(x)
    file2.close()

    raw_data = pd.read_csv(flname)
    return raw_data
raw_data = raw_data_exange(11)[1:-1]
raw_data = raw_data[raw_data['back'] != 'None'].reset_index()
raw_data = raw_data.drop(columns=['Mtext','Ftext','hint','index'])
# raw_data.head()
# raw_data.to_csv(flname)
```

```python
def data_cut(df, back, emotion):
    df = df[df['emotion'] == emotion]
    df = df[df['back'] == back]

    df = df[df['corrAns'] != 'None' ]  # delete practice data or instruct items
    df = df.dropna(subset=['key_2.corr'])

    return df
r_data = data_cut(raw_data, '2-back', 'negative')
r_data
```
