```python
import pandas as pd
import numpy as np
import os

pds = pd.read_excel("neu_1.xlsx")
pd2 = pd.read_excel("neu_2.xlsx")
pd3 = pd.read_excel("neu_3.xlsx")
pd4 = pd.read_excel("practice_neu_1.xlsx")
pd5 = pd.read_excel("practice_neu_2.xlsx")
pd6 = pd.read_excel("practice_neu_3.xlsx")
# pds = pds.append(pd2, pd3)
c = pds.Pic
c = c.append(pd3.Pic).append(pd2.Pic).append(pd4.Pic).append(pd5.Pic).append(pd6.Pic)
c
```

```python
dir_neg = "source/neutral/"


files = os.listdir(dir_neg)
full_paths = []
# 遍历文件列表，使用 os.path.join() 函数将文件名和目录名合并为完整的路径
for filename in files:
    # 使用 os.path.join() 函数将目录名和文件名合并为完整的路径
    full_path = os.path.join(dir_neg, filename)
    full_paths.append(full_path)
print(len(full_paths))
full_paths
```

```python
def check_and_del(A,B):
    for file_path in B:
        if not os.path.exists(file_path):
            continue
        if file_path not in A:
            os.remove(file_path)
            print('删除文件：', file_path)
```

```python
# 判断B集合是不是在A集合中，如果不在的全部删除


A = set(c)
B = set(full_paths)


check_and_del(A,B)
```
