```python
import pandas as pd
import numpy as np 
from scipy import stats
```

```python
def T_test(nplist_1, nplist_2):
    return stats.ttest_rel(nplist_1, nplist_2)
```

```python
data = pd.read_excel('CFAPS_P.XLS')
data
```

```python
NF=data[data['编号'].str.startswith('NF')][61:80].reset_index()
NM=data[data['编号'].str.startswith('NM')][61:80].reset_index()
PF=data[data['编号'].str.startswith('PF')][1:60].reset_index()
PM=data[data['编号'].str.startswith('PM')][1:60].reset_index()
TF=data[data['编号'].str.startswith('TF')][1:60].reset_index()
TM=data[data['编号'].str.startswith('TM')][1:60].reset_index()
NM
```

```python
# 积极目标 vs 消极分心
flag = True
while flag:
    pf = PF.sample(3)
    pm = PM.sample(3)
    nf = NF.sample(3)
    nm = NM.sample(3)

    df = pd.concat([pf,pm,nf,nm])
    df.rename(columns={'编号':'number','愉悦度':'pleasantness','唤醒度':'arousal','优势度':'dominance','吸引度':'attraction'},inplace=True)
    corrAns = []
    location = []
    for label in df['number'].values:
        if label.startswith('P'):
            corrAns.append('space')
            if label.startswith('PF'):
                loca = 'source/positive/FEMALE/'+label.lstrip('PF')+'.GIF'
            else:
                loca = 'source/positive/MALE/'+label.lstrip('PM')+'.GIF'
            location.append(loca)
        elif label.startswith('N'):
            corrAns.append('None')
            if label.startswith('NF'):
                loca = 'source/negative/FEMALE/'+label.lstrip('NF')+'.GIF'
            else:
                loca = 'source/negative/MALE/'+label.lstrip('NM')+'.GIF'
            location.append(loca)

    df['corrAns'] = corrAns
    df['location'] = location
    df['cond'] = 'htsd'
    df.reset_index()


    tar = df[df['corrAns'] == 'space']
    dis = df[df['corrAns']!='space']

    if T_test(tar['pleasantness'],dis['pleasantness']).pvalue < 0.05 and T_test(tar['arousal'],dis['arousal']).pvalue > 0.05:
        flag = False
        print("\t")

        PF = pd.concat([PF,pf,pf]).drop_duplicates(keep=False)
        PM = pd.concat([PM,pm,pm]).drop_duplicates(keep=False)
        NF = pd.concat([NF,nf,nf]).drop_duplicates(keep=False)
        NM = pd.concat([NM,nm,nm]).drop_duplicates(keep=False)

print("Pleasantness: ",T_test(tar.pleasantness, dis.pleasantness))
print("Arousal: ",T_test(tar.arousal, dis.arousal))
df
# len(NM)
```

```python
# 写数据
df.to_excel('trial_A.xlsx',index=False)
```
